```python
from PIL import Image
```

```python
def to_binary(value):
    binary = ""
    while value > 0:
        binary = str(value % 2) + binary
        value = value // 2
    return binary.zfill(8)

def cellular_automaton(rule, width, height):
    # Inicializa√ß√£o da matriz do aut√¥mato celular
    automaton = [[0] * width]
    automaton[0][width // 2] = 1  # Definindo a c√©lula central como 1

    # Defini√ß√£o das regras
    ruleset = to_binary(rule)

    # Gera√ß√£o das pr√≥ximas linhas do aut√¥mato celular
    for i in range(1, height):
        row = []
        for j in range(width):
            # Obtendo os valores das c√©lulas vizinhas
            left = automaton[i - 1][(j - 1) % width]
            center = automaton[i - 1][j]
            right = automaton[i - 1][(j + 1) % width]

            # Convertendo os valores das c√©lulas vizinhas em uma sequ√™ncia bin√°ria
            pattern = ''.join([str(left), str(center), str(right)])

            # Aplicando as regras
            index = 7 - int(pattern, 2)  # Revertendo o padr√£o bin√°rio para corresponder √† ordem das regras
            next_cell = int(ruleset[index])

            row.append(next_cell)

        automaton.append(row)

    # Imprimindo o aut√¥mato celular como uma figura
    for row in automaton:
        line = ''.join(['#' if cell == 1 else ' ' for cell in row])
        print(line)

    # Criando a imagem
    image = Image.new("1", (width, height))

    # Preenchendo a imagem com os valores do aut√¥mato celular
    for y, row in enumerate(automaton):
        for x, cell in enumerate(row):
            pixel = 255 if cell == 1 else 0
            image.putpixel((x, y), pixel)

    # Salvando a imagem como um arquivo PNG
    image.save("automaton.png")

    print("Imagem salva com sucesso!")


# Par√¢metros iniciais
L = 80
H = 100
rule = int(input("Digite o n√∫mero da regra (0-255): "))

# Executando o aut√¥mato celular
cellular_automaton(rule, L, H)
```
