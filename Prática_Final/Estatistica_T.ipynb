```python
import pandas as pd
from scipy.stats import f_oneway
```

```python
#Import data
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pratica_Final_MMC/ohlc1d.csv")
```

```python
# Selecionar as colunas 'Open' e 'Close'
open_data = data['Open']
close_data = data['Close']

# Executar o teste F
f_statistic, p_value = f_oneway(open_data, close_data)

# Imprimir o resultado do teste
alpha = 0.05  # Nível de significância
print("Resultado do teste F:")
print("Estatística F:", f_statistic)
print("Valor-p:", p_value)
if p_value < alpha:
    print("Rejeitar a hipótese nula. As variáveis não são estatisticamente iguais.")
else:
    print("Não há evidências para rejeitar a hipótese nula. As variáveis são estatisticamente iguais.")
```
