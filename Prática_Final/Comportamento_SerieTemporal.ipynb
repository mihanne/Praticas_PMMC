```python
import numpy as np
import os
import pandas as pd
from pandas import DataFrame
from pandas import Series
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
#Import data
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pratica_Final_MMC/ohlc3d2023.csv",header=0, index_col='Datetime')
```

```python
data['Open'].fillna(method='ffill', inplace=True)
data['High'].fillna(method='ffill', inplace=True)
data['Low'].fillna(method='ffill', inplace=True)
data['Close'].fillna(method='ffill', inplace=True)
print (data)
```

```python
s1 = pd.Series(data['Open'])
data['Media'] = s1.rolling(30).mean()
data['Desvio_Padrao'] = s1.rolling(30).std()
```

```python
data[['Open', 'Media', 'Desvio_Padrao']].plot(figsize=(10, 8))
plt.title('Bitcoin - Valor de Abertura (Open) com variação de um 3 dias')
plt.ylabel('Preço $')
plt.xlabel('Date')
plt.show()
```

```python
s1 = pd.Series(data['Close'])
data['Media'] = s1.rolling(30).mean()
data['Desvio_Padrao'] = s1.rolling(30).std()
```

```python
data[['Close', 'Media', 'Desvio_Padrao']].plot(figsize=(10, 8))
plt.title('Bitcoin - Valor de Fechamento (Close) com variação de um 3 dias')
plt.ylabel('Preço $')
plt.xlabel('Date')
plt.show()
```
