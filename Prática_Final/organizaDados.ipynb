```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
#Import data
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pratica_Final_MMC/bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv")
#Transform timestamp to datetime.
data['Datetime'] = pd.to_datetime(data['Timestamp'], unit='s')
```

```python
# Prepare DataFrame
data = data.set_index('Datetime')
data = data.drop(["Timestamp"], axis=1)
```

```python
# Resample data to 180-minute intervals
df_ohlc = data.resample('14400T').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last'})
```

```python
# Forward fill missing values
df_ohlc = df_ohlc.ffill()
```

```python
# Save resampled data to CSV
df_ohlc.to_csv('/content/drive/MyDrive/Colab Notebooks/Pratica_Final_MMC/ohlc10d.csv')
print(df_ohlc)
```
