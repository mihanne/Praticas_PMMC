```python
import matplotlib.pyplot as plt
```

```python
# Definindo as variáveis iniciais
V = float(input("Velocidade inicial (m/s): "))
H = float(input("Altura inicial (m): "))
g = 9.81 # aceleração da gravidade em m/s^2
dt = 0.01 # passo de tempo em segundos
imax = 10000 # número máximo de passos
```

```python
# Inicializando as variáveis de tempo, posição e velocidade
ti = t0 = 0
xi = x0 = H
vi = v0 = V
```

```python
# Criando listas para armazenar os valores de tempo, posição e velocidade
t_list = [t0]
x_list = [x0]
v_list = [v0]
```

```python
# Loop para calcular a posição e velocidade em cada passo de tempo
for i in range(1, imax+1):
    # Atualizando o tempo
    ti = ti + dt
    
    # Calculando a aceleração, posição e velocidade utilizando o método de Euler
    a = -g
    xi = xi + vi*dt + 0.5*a*dt**2
    vi = vi + a*dt
    
    # Adicionando os valores de tempo, posição e velocidade às listas
    t_list.append(ti)
    x_list.append(xi)
    v_list.append(vi)
    
    # Verificando se a partícula atingiu o solo
    if xi <= 0:
        xi = 0
        break
```

```python
# Plotando o gráfico de altura em função do tempo
plt.subplot(311)
plt.plot(t_list, x_list)
plt.title("Queda Livre")
plt.ylabel("Altura (m)")

# Plotando o gráfico de velocidade em função do tempo
plt.subplot(312)
plt.plot(t_list, v_list)
plt.ylabel("Velocidade (m/s)")

# Calculando a aceleração em cada passo de tempo
a_list = [-g] * len(t_list)
for i in range(len(t_list)):
    if i == 0:
        continue
    else:
        a_list[i] = (v_list[i] - v_list[i-1])/dt

# Plotando o gráfico de aceleração em função do tempo
plt.subplot(313)
plt.plot(t_list, a_list)
plt.xlabel("Tempo (s)")
plt.ylabel("Aceleração (m/s^2)")

# Exibindo o gráfico
plt.show()
```
